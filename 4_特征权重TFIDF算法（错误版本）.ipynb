```python
def readnegfile(i):
    text=open('neg.'+str(i)+'.txt').read()
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

if __name__ == '__main__':
    list_text=[]
    for i in range(11):
        list_text.append(readnegfile(i))
       
    vectorizer = TfidfVectorizer()
    vectorizer.fit(list_text)
    tfidf_text = vectorizer.transform(list_text)
    print (tfidf_text.toarray())
```
